# FEATURE ENGINEERING 

Now let's feature engineer some new features based off of our given features and are relevant to our domain!

````python

def regions(x):
  x=x.upper()
  new_england = ['CONNECTICUT', 'MAINE', 'MASSACHUSETTS', 'NEW_HAMPSHIRE', 'RHODE_ISLAND', 'VERMONT']
  mideast = ['DELAWARE', 'DISTRICT_OF_COLUMBIA', 'MARYLAND', 'NEW_JERSEY', 'NEW_YORK', 'PENNSYLVANIA']
  great_lakes = ['ILLINOIS', 'INDIANA', 'MICHIGAN', 'OHIO', 'WISCONSIN']
  plains = ['IOWA', 'KANSAS', 'MINNESOTA', 'MISSOURI', 'NEBRASKA', 'NORTH_DAKOTA', 'SOUTH_DAKOTA']
  southeast = ['ALABAMA', 'ARKANSAS', 'FLORIDA', 'GEORGIA', 'KENTUCKY', 'LOUISIANA', 'MISSISSIPPI', 'NORTH_CAROLINA', 'SOUTH_CAROLINA', 'TENNESSEE', 'VIRGINIA', 'WEST_VIRGINIA']
  southwest = ['ARIZONA', 'NEW_MEXICO', 'OKLAHOMA', 'TEXAS']
  rocky_mnt = ['COLORADO', 'IDAHO', 'MONTANA', 'UTAH', 'WYOMING']
  far_west = ['ALASKA', 'CALIFORNIA', 'HAWAII', 'NEVADA', 'OREGON', 'WASHINGTON']
  
  if (x in new_england):
    return 'NEW_ENGLAND'
  elif (x in mideast):
    return 'MIDEAST'
  elif (x in great_lakes):
    return 'GREAT_LAKES'
  elif (x in plains):
    return 'PLAINS'
  elif (x in southeast):
    return 'SOUTHEAST'
  elif (x in southwest):
    return 'SOUTHWEST'
  elif (x in rocky_mnt):
    return 'ROCKY MOUNTAIN'
  else:
    return 'FAR WEST'

final_data['BEA_REGIONS'] = final_data['STATE'].apply(regions)

  
````

After some investigation, we found a slight descrepancy present within the data. It seems that the 

```
TOTAL_EXPENDITURE
```
feature does not equal the sum of the 

```
INSTRUCTION_EXPENDITURE
SUPPORT_SERVICES_EXPENDITURE 
OTHER_EXPENDITURE
CAPITAL_OUTLAY_EXPENDITURE 
```
features. This is most likely due to how this dataset is a result of other datasets combined, including Census data, which contains large number of  specific features that may have been missed by this dataset's author. Thus, due to this reality-check, we will drop the 


```
TOTAL_EXPENDITURE 
```
feature. 




````python
# df['boolean_column'] = np.where(df['some_column'] > 15, True, False)

final_data['Add Up'] = np.where(final_data['TOTAL_EXPENDITURE'] > final_data['INSTRUCTION_EXPENDITURE'] + final_data['SUPPORT_SERVICES_EXPENDITURE'] + final_data['OTHER_EXPENDITURE']
       + final_data['CAPITAL_OUTLAY_EXPENDITURE'], True, False)
````

````python
final_data['Add Up'].value_counts()
````

````python
del final_data['Add Up']
del final_data['TOTAL_EXPENDITURE']
````
